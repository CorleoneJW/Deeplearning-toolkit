```python
import numpy as np
import matplotlib.pyplot as plt

# groups and labels for groups
groups = ['FLARE26', 'TCIA', 'AMOS','Synapse']
subgroups = ['client_BC', 'w/o A', 'client_AC', 'w/o B', 'client_AB', 'w/o C', 'client_ABC']

# data of the first subplot
data1 = np.array([
    [54.93, 87.13, 71.39, 85.05, 83.37, 85.99, 84.77],
    [49.73, 72.65, 66.56, 75.46, 52.14, 73.59, 72.34],
    [60.53, 64.63, 55.37, 62.13, 52.40, 63.60, 66.95],
    [68.39, 77.54, 59.62, 78.66, 69.75, 75.14, 78.45]
])

# data of the second subplot
data2 = np.array([
    [80.0, 85.0, 87.0, 89.0, 90.0, 91.0, 92.0],
    [55.0, 65.0, 60.0, 63.0, 65.0, 70.0, 72.0],
    [58.0, 62.0, 57.0, 56.0, 50.0, 63.0, 61.0],
    [75.0, 80.0, 77.0, 84.0, 72.0, 78.0, 82.0]
])

color = ['#DDF2C6', '#C0D1DC', '#C8F2E9', '#CFCCE3', '#AED4CB', '#BBA8BF', '#F19C99']

# set the size of the figure
plt.figure(figsize=(24, 6))

# create first subplot
plt.subplot(1, 2, 1)

# set the width and gap of bars
bar_width = 0.1
group_spacing = 0.4
subgroup_spacing = 0.02

# calculate the position of each group
index = np.arange(len(groups))

# calculate the position of each sub group
subgroup_indices = (index[:, np.newaxis] +
                    (bar_width + subgroup_spacing) * np.arange(len(subgroups)))

# align the group1 and group2
subgroup_indices[:, 0] += subgroup_spacing
subgroup_indices[:, 2] += subgroup_spacing
subgroup_indices[:, 4] += subgroup_spacing
subgroup_indices[:, 6] += subgroup_spacing

# draw the bar chart
for i, subgroup in enumerate(subgroups):
    subgroup_data = data1[:, i]
    plt.bar(subgroup_indices[:, i], subgroup_data, bar_width, label=subgroup, color=color[i])

# set the labels of x and y axis
plt.xlabel('Stomach', fontsize=26)
plt.ylabel('Dice Value (%)', fontsize=24)

# set the ticks for x and y
plt.xticks(index + group_spacing, groups, fontsize=26)
plt.yticks(fontsize=24)

# create the second subplot
plt.subplot(1, 2, 2)

# Using different data to draw the second subplot
for i, subgroup in enumerate(subgroups):
    subgroup_data = data2[:, i]
    plt.bar(subgroup_indices[:, i], subgroup_data, bar_width, label=subgroup, color=color[i])

# set the labels for x and y axis
plt.xlabel('Pancreas', fontsize=26)
plt.ylabel('Dice Value (%)', fontsize=24)

# set the ticks for x and y axis
plt.xticks(index + group_spacing, groups, fontsize=24)
plt.yticks(fontsize=24)

# adjust tht distance from each other
plt.tight_layout()

# add the legned for whole chart
handles, labels = plt.gca().get_legend_handles_labels()
plt.figlegend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=len(subgroups), fontsize=24)

# show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# groups and labels for groups
groups = ['FLARE26', 'TCIA', 'AMOS', 'Synapse']
subgroups = ['client_23', 'P_1', 'client_13', 'P_2', 'client_12', 'P_3', 'client_123']

# data of the first subplot
data1 = np.array([
    [54.93, 87.13, 71.39, 85.05, 83.37, 85.99, 84.77],
    [49.73, 72.65, 66.56, 75.46, 52.14, 73.59, 72.34],
    [60.53, 64.63, 55.37, 62.13, 52.40, 63.60, 66.95],
    [68.39, 77.54, 59.62, 78.66, 69.75, 75.14, 78.45]
])

# data of the second subplot
data2 = np.array([
    [80.0, 85.0, 87.0, 89.0, 90.0, 91.0, 92.0],
    [55.0, 65.0, 60.0, 63.0, 65.0, 70.0, 72.0],
    [58.0, 62.0, 57.0, 56.0, 50.0, 63.0, 61.0],
    [75.0, 80.0, 77.0, 84.0, 72.0, 78.0, 82.0]
])

color = ['#DDF2C6', '#C0D1DC', '#C8F2E9', '#CFCCE3', '#AED4CB', '#BBA8BF', '#F19C99']

# Create subplots with two rows and one column
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12), sharex=False)

# set the width and gap of bars
bar_width = 0.1
group_spacing = 0.4
subgroup_spacing = 0.02

# calculate the position of each group
index = np.arange(len(groups))

# calculate the position of each sub group
subgroup_indices = (index[:, np.newaxis] +
                    (bar_width + subgroup_spacing) * np.arange(len(subgroups)))

# align the group1 and group2
subgroup_indices[:, 0] += subgroup_spacing
subgroup_indices[:, 2] += subgroup_spacing
subgroup_indices[:, 4] += subgroup_spacing
subgroup_indices[:, 6] += subgroup_spacing

"""subplot 1"""

# draw the bar chart for the first subplot
for i, subgroup in enumerate(subgroups):
    subgroup_data = data1[:, i]
    ax1.bar(subgroup_indices[:, i], subgroup_data, bar_width, label=subgroup, color=color[i])

# set the labels of y axis for the first subplot
ax1.set_xlabel('Stomach', fontsize=26)
ax1.set_ylabel('Dice Value (%)', fontsize=24)

# set the x ticks and labels for the first subplot
ax1.set_xticks(index + group_spacing)
ax1.set_xticklabels(groups, fontsize=24)

"""subplot 2"""

# draw the bar chart for the second subplot
for i, subgroup in enumerate(subgroups):
    subgroup_data = data2[:, i]
    ax2.bar(subgroup_indices[:, i], subgroup_data, bar_width, label=subgroup, color=color[i])

# set the labels of x and y axis for the second subplot
ax2.set_xlabel('Pancreas', fontsize=26)
ax2.set_ylabel('Dice Value (%)', fontsize=24)

# set the x ticks and labels for the second subplot
ax2.set_xticks(index + group_spacing)
ax2.set_xticklabels(groups, fontsize=24)

"""public part"""

# set the ticks for y axis for both subplots
ax1.yaxis.set_tick_params(labelsize=24)
ax2.yaxis.set_tick_params(labelsize=24)

# add the legend for the whole chart to the right with adjusted position
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

# Combine the handles and labels from both subplots
all_handles = handles1
all_labels = labels1

# Place the legend to the right of the second subplot
legend = fig.legend(all_handles, all_labels, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=24, frameon=False)

# adjust the layout
fig.tight_layout(h_pad=3)  # adjust parameter h_pad to change the gap between plots

# show the plot
plt.show()
```
